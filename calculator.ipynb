```python
operator = input("Enter Operator: ")
raw_val1 = input("Enter val 1: ")
raw_val2 = input("Enter val 2: ")

# Check if the user actually typed something
if raw_val1 != "" and raw_val2 != "":
    val1 = int(raw_val1)
    val2 = int(raw_val2)

    if operator == "+":
        print(val1 + val2)
    elif operator == "-":
        print(val1 - val2)
    elif operator == "*":
        print(val1 * val2)
    elif operator == "/":
        if val2 != 0:
            print(val1 / val2)
        else:
            print("Error: Cannot divide by zero!")
    else:
        print("Invalid operation!")
else:
    print("Both values are required")
```
