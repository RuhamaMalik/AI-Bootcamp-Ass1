```python
# Qno : 7

maths = int(input("Maths marks: "))
english = int(input("English marks: "))
urdu = int(input("Urdu marks: "))
physics = int(input("Physics marks: "))
islamiat = int(input("Islamiat marks: "))

total = maths + english + urdu + physics + islamiat
percent = (total / 500) * 100

print(f"\nTotal Marks: {total}/500")
print(f"Percentage: {percent}%")

if percent >= 80:
    print("Grade: A+")
elif percent >= 70:
    print("Grade: A")
elif percent >= 60:
    print("Grade: B")
elif percent >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```
